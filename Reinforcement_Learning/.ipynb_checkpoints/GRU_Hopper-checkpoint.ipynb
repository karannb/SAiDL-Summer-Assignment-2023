```python
from model_ import *
from expt_ import *
from trainer_ import *
import torch
import numpy as np
import matplotlib.pyplot as plt
import math

np.random.seed(1)
torch.manual_seed(1)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(30, 15))
epochs = 10
Experiment1 = Expt(ax1, ax2, mode='normal', dataset='expert', iterations=100, epochs=epochs, cell_type='GRU')
```

```python
Experiment1.experiment()
legend = [f"epoch{math.ceil((num+1)/2)}_{3600 if num%2==0 else 1800}" for num in range(epochs*2)]
ax1.legend(legend, prop={"size":20})
ax2.legend(legend, prop={"size":20})
fig
```
