# 6. Dice score as Loss -

```python
class DICELoss(nn.Module): #TverskyLoss
    def __init__(self, alpha=0.7, smooth=1):
        super(DICELoss, self).__init__()
        
        self.alpha = alpha
        self.smooth = smooth
        
    def forward(self, y_pred, y_true):
        
        #y_pred.squeeze(1)
        y_pred_pos = torch.sigmoid(y_pred)
        
        y_pred_pos = y_pred_pos.view(-1)
        y_true = y_true.view(-1)
        
        TP = (y_pred_pos * y_true).sum()    
        #FP = ((1-y_true) * y_pred_pos).sum()
        #FN = (y_true * (1-y_pred_pos)).sum()
       
        dice_score = (2*TP + self.smooth) / (2*TP + y_pred_pos.sum() + y_true.sum() + self.smooth)
        
        return 1 - dice_score
```

# 7. Focal Loss -

```python
class FocalLoss(nn.Module):
    
    def __init__(self, alpha=0.8, gamma=2):
        super(FocalLoss, self).__init__()
        self.gamma = gamma
        self.alpha = alpha #hyperparameter, balancing factor, gives more weight to 1s if set properly
        
    def forward(self, y_pred, y_true):
        
        pred = torch.sigmoid(y_pred)
        
        pred = pred.view(-1)
        y_true = y_true.view(-1)
        
        cross_entropy = F.binary_cross_entropy(pred, y_true, reduction='mean')
        exp = torch.exp(cross_entropy)
        #alpha = y_true*self.alpha + (1. - y_true)*(1. - self.alpha)
        #pt = torch.where(y_true==1., pred, 1.-pred)
        focal_loss = self.alpha*((1-exp)**self.gamma)*cross_entropy
        
        return focal_loss
```
